### Read from Silver

```python
customers_df = spark.table("olist_ecommerce.silver.customers")
```

```python
dim_customers_df = (
    customers_df
    .select(
        "customer_id",
        "customer_unique_id",
        "zip_code",
        "city",
        "state"
    )
    .dropDuplicates(["customer_id"])
)
```

### Write to Gold

```python
dim_customers_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("olist_ecommerce.gold.dim_customers")
```
