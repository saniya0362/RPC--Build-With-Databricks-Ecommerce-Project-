### Read from Silver

```python
from pyspark.sql import functions as F

SILVER_DB = "olist_ecommerce.silver"

products_df = spark.table(f"{SILVER_DB}.products")
category_df = spark.table(f"{SILVER_DB}.product_category_en")
```

```python
dim_products_df = (
    products_df
    .join(
        category_df,
        "product_category_name",
        "left"
    )\
    .withColumnRenamed("category_name","product_category_name_english")\
    .select(
        "product_id",
        "product_category_name_english",
        "product_weight_g",
        "product_length_cm",
        "product_height_cm",
        "product_width_cm",
        "product_volume_cm3"
    )
)
```

### Write to Gold

```python
dim_products_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("olist_ecommerce.gold.dim_products")
```
