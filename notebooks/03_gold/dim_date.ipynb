### Read Silver tables

```python
from pyspark.sql import functions as F

orders_df = spark.table("olist_ecommerce.silver.orders")
```

```python
dim_date_df = (
    orders_df
    .select(F.to_date("purchase_timestamp").alias("date"))
    .distinct()
    .withColumn("year", F.year("date"))
    .withColumn("month", F.month("date"))
    .withColumn("day", F.dayofmonth("date"))
    .withColumn("week_of_year", F.weekofyear("date"))
    .withColumn("day_of_week", F.date_format("date", "E"))
)
```

### Write to Gold

```python
dim_date_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("olist_ecommerce.gold.dim_date")
```
