### Create Catalog

```python
%sql
CREATE CATALOG olist_ecommerce;
```

### Create Schemas

```python
%sql
USE CATALOG olist_ecommerce;

-- Bronze schema
CREATE SCHEMA IF NOT EXISTS olist_ecommerce.bronze;

-- Silver schema
CREATE SCHEMA IF NOT EXISTS olist_ecommerce.silver;

-- Gold schema
CREATE SCHEMA IF NOT EXISTS olist_ecommerce.gold;
```

```python
%sql
SHOW SCHEMAS IN olist_ecommerce;
```
